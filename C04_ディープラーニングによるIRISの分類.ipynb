```python
from keras.models import Sequential
from keras.layers.convolutional import Conv2D
from keras.layers.pooling import MaxPool2D
from keras.optimizers import Adam

from keras.layers.core import Dense, Activation, Dropout, Flatten
from keras.utils import plot_model
from keras.callbacks import TensorBoard

from keras.datasets import cifar10
from keras.utils import np_utils
```

```python
#irisに関するデータの読み込み
   # ３品種　４つの特徴量　サンプル数150個
   # Keras　TensorFrow を使用する
    
from sklearn.datasets import load_iris
iris_dataset = load_iris()

print("Keys of iris_dataset: {}".format(iris_dataset.keys()))
print('  ')
print(iris_dataset['DESCR'][:193] + "\n...")
print('  ')
print("Feature names: {}".format(iris_dataset['feature_names']))
print('  ')
print("Target names: {}".format(iris_dataset['target_names']))
print('  ')
print("First five rows of data:\n{}".format(iris_dataset['data'][:5]))
print('  ')
print("First five rows of target:\n{}".format(iris_dataset['target'][:5]))
```

```python

#　サンプルをランダムに学習用データとテスト用データに分ける
   # X:特徴量ベクトル　　y:　品種

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    iris_dataset['data'], iris_dataset['target'], random_state=0)

# 
print("X_train shape: {}".format(X_train.shape))
print("y_train shape: {}".format(y_train.shape))
print('  ')
print("X_test shape: {}".format(X_test.shape))
print("y_test shape: {}".format(y_test.shape))
print('  ')
print("First five rows of data:\n{}".format(X_train[:5]))
print('  ')
print("First five rows of target:\n{}".format(y_train[:5]))

```

```python
#　CNNの出力データ形式（one-hotベクトル）に変更する

from keras.utils.np_utils import to_categorical

yc_train = to_categorical(y_train)#　出力値をone_hot形式に変換
yc_test  = to_categorical(y_test)
print("one_hot of target:\n{}".format(yc_train[:5]))
```

```python
#モデル１　　

#　ディープラーニングのモデル（レイヤ・ノード構造）を構築する
#　隠れ層　1層（ノード数8個）
import keras
from keras.layers import Activation, Dense
from keras.layers.normalization import BatchNormalization
from keras.models import Sequential
from keras.utils import plot_model

model = Sequential()
model.add(Dense (units=8,input_dim= 4 ))#　入力層と中間層のノード数を定義
model.add(Activation('relu'))#活性化関数を定義

model.add(Dense(3))#出力層のノード数を定義
model.add(Activation('softmax'))

# モデルをプロット
plot_model(model,show_shapes=True, to_file='model.png' )
model.summary()

```

```python
#　モデルのコンパイル、学習
model.compile(optimizer='adam',
             loss='categorical_crossentropy',
             metrics=['accuracy'])

model.fit(X_train,yc_train,epochs=20,batch_size=32)

#　評価
_, acc = model.evaluate(X_test, yc_test, verbose=0)
print('  ')
print('accuracy: {:.2f}'.format(acc))

scores = model.evaluate(X_test, yc_test, verbose=0)
print("%s: %.2f%%" % (model.metrics_names[1], scores[1] * 100))

```

```python
#モデル２

#　ディープラーニングのモデル（レイヤ・ノード構造）を構築する
#　隠れ層1層（ノード数16個）
import keras
from keras.layers import Activation, Dense
from keras.layers.normalization import BatchNormalization
from keras.models import Sequential
from keras.utils import plot_model

model = Sequential()
model.add(Dense (units=16,input_dim= 4 ))#　入力層と中間層のノード数を定義
model.add(Activation('relu'))#活性化関数を定義

model.add(Dense(3))#出力層のノード数を定義
model.add(Activation('softmax'))

# モデルをプロット
model.summary()

```

```python
#　モデルのコンパイル、学習
model.compile(optimizer='adam',
             loss='categorical_crossentropy',
             metrics=['accuracy'])

model.fit(X_train,yc_train,epochs=100,batch_size=32)

#　評価
_, acc = model.evaluate(X_test, yc_test, verbose=0)
print('  ')
print('accuracy: {:.2f}'.format(acc))

scores = model.evaluate(X_test, yc_test, verbose=0)
print("%s: %.2f%%" % (model.metrics_names[1], scores[1] * 100))
```

```python
#モデル３

#　ディープラーニングのモデル（レイヤ・ノード構造）を構築する
#　隠れ層2層（ノード数8個+ノード数16）

import keras
from keras.layers import Activation, Dense
from keras.layers.normalization import BatchNormalization
from keras.models import Sequential
from keras.utils import plot_model

model = Sequential()
model.add(Dense (units=8,input_dim= 4 ))#　入力層と中間層のノード数を定義
model.add(BatchNormalization())
model.add(Activation('relu'))#活性化関数を定義

model.add(Dense(16))
model.add(BatchNormalization())
model.add(Activation('relu'))

model.add(Dense(3))#出力層のノード数を定義
model.add(Activation('softmax'))

# モデルをプロット
model.summary()


from IPython.display import SVG
from keras.utils.vis_utils import model_to_dot

```

```python
#　モデルのコンパイル、学習
model.compile(optimizer='adam',
             loss='categorical_crossentropy',
             metrics=['accuracy'])

model.fit(X_train,yc_train,epochs=200,batch_size=32)

#　評価
_, acc = model.evaluate(X_test, yc_test, verbose=0)
print('  ')
print('accuracy: {:.2f}'.format(acc))

scores = model.evaluate(X_test, yc_test, verbose=0)
print("%s: %.2f%%" % (model.metrics_names[1], scores[1] * 100))

```
